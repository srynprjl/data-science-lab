# Question 1
- Import pandas and create a DataFrame with columns: Name, Age, Marks.
Display the first 5 rows and dataset shape.

```python
import pandas as pd
import random

data = {
    "Name": ["Aabhash", "Sandeep", "Shreyan", "Amit", "Yash", "Anushka"],
    "Age": [20, 21, 19, 22, 20, 21],
    "Marks": [85, 55, 92, 40, 78, 95],
}

df = pd.DataFrame(data)
# df.to_csv("data.csv", index=False)
print("Data:\n", df.head())
print("\nShape:", df.shape)
```

# Question 2

- Load a CSV file into a DataFrame.
- Display column names, data types, and basic statistics.

```python
csv_df = pd.read_csv("data.csv")
print("Columns:", csv_df.columns)
print("\nDatatype:\n", csv_df.dtypes)
print("\nStats:\n", csv_df.describe())
```

# Question 3

- Select rows where Marks > 60.
- Select only Name and Marks columns.

```python
df = csv_df.copy()
high_marks = df[df["Marks"] > 60]
subset = df[["Name", "Marks"]]

print(f"High Marks:\n {high_marks} \n")
print(f"Subset:\n {subset} \n")
```

# Question 4

- Check for missing values in the dataset.
- Fill missing numerical values with the mean.

```python
print("All Null Values:\n", df.isnull().sum())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
print(df.head(6))
```

# Question 5
- Detect duplicate rows in the dataset.
- Remove duplicates and verify the result.

```python
print(f"Duplicate Rows (After): \n {df.duplicated().sum()} \n")
df.drop_duplicates(inplace=True)
print(f"Duplicate Rows (Before): \n {df.duplicated().sum()} \n")
```

# Question 6

- Identify outliers in the Marks column using the IQR method.
- Remove the outliers from the dataset.


```python
q1 = df["Marks"].quantile(0.25)
q3 = df["Marks"].quantile(0.75)
iqr = q3 - q1

low = q1 - 1.5 * iqr
up = q3 + 1.5 * iqr
df = df[(df["Marks"] >= low) & (df["Marks"] <= up)]
print(f"Low: {low}, High: {up}")

print(df)
```

# Question 7
- Create a new column by transforming Marks (e.g., Marks / 10).
- Rename columns and save the cleaned dataset to a CSV file.

```python
df["Transformed Marks"] = df["Marks"]/10
print(f"Transformed Marks: \n {df}")
df.rename(columns={"Transformed Marks": "New Marks"}, inplace=True)
print(f"\nRenamed Column: \n {df}")

df.to_csv("cleaned_data.csv", index=False)
```
